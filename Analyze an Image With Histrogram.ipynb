```python
import cv2
import matplotlib.pyplot as plt
img1=cv2.imread(r"R:\Ruponty\IMG_20250328_200825.jpg")
img2=cv2.imread(r"R:\Ruponty\IMG_20250328_200543.jpg")
img3=cv2.imread(r"R:\Ruponty\IMG_20250328_194340.jpg")
img1=cv2.resize(img1,(300,300))
img2=cv2.resize(img2,(300,300))
img3=cv2.resize(img3,(300,300))
hist1=cv2.calcHist([img1],[0],None,[255],[0,255])#image,colorChannel,mask,pixelnumer,range
hist2=cv2.calcHist([img2],[0],None,[255],[0,255])
hist3=cv2.calcHist([img3],[0],None,[255],[0,255])
plt.figure(figsize=(30,30))
plt.subplot(3,1,1)
plt.plot(hist1)
plt.show()
plt.figure(figsize=(30,30))
plt.subplot(3,1,2)
plt.plot(hist2)
plt.show()
plt.figure(figsize=(30,30))
plt.subplot(3,1,3)
plt.plot(hist3)
plt.show()
cv2.imshow("apu1",img1)
cv2.imshow("apu2",img2)
cv2.imshow("apu3",img3)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
import matplotlib.pyplot as plt
img1=cv2.imread(r"R:\Ruponty\IMG_20250328_200825.jpg")

img1=cv2.resize(img1,(300,300)) 
b,g,r=cv2.split(img1)

plt.hist(b.ravel(),255,(0,255))
plt.hist(g.ravel(),255,(0,255))
plt.hist(r.ravel(),255,(0,255))
plt.show()

cv2.imshow("apu3",img1)
cv2.imshow("apu1",b)
cv2.imshow("apu2",g)
cv2.imshow("apu3",r)

cv2.waitKey(0)
cv2.destroyAllWindows()
```
