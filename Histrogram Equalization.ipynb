```python
import cv2
import matplotlib.pyplot as plt

i=cv2.imread(r"R:\Mobile's Photo\DCIM\Camera\IMG_20240225_142741.jpg")
im=cv2.resize(i,(500,500))
g=cv2.cvtColor(im,cv2.COLOR_BGR2GRAY)
eq=cv2.equalizeHist(g)

hist=cv2.calcHist([g],[0],None,[255],[0,255])
hist1=cv2.calcHist([eq],[0],None,[255],[0,255])

plt.plot(hist)
plt.plot(hist1,c="red")
plt.show()

cv2.imshow("apu",g)
cv2.imshow("apu1",eq)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
import matplotlib.pyplot as plt

i=cv2.imread(r"R:\ddca3f92-4b8e-4672-bb6b-f3594ad4e304.jpg")
im=cv2.resize(i,(500,500))
g=cv2.cvtColor(im,cv2.COLOR_BGR2GRAY)
eq=cv2.equalizeHist(g)
clahe=cv2.createCLAHE(clipLimit=1,tileGridSize=(8,8))
f=clahe.apply(g)

hist=cv2.calcHist([g],[0],None,[255],[0,255])
hist1=cv2.calcHist([eq],[0],None,[255],[0,255])


plt.plot(hist)
plt.plot(hist1,c="red")
plt.plot(f,c="green")
plt.show()

cv2.imshow("apu",g)
cv2.imshow("apu1",eq)
cv2.imshow("apu2",f)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
