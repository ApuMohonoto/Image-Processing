```python
import cv2

i1 = cv2.imread(r"R:\stock.jpg")
i = cv2.resize(i1, (500, 500))


g = cv2.cvtColor(i, cv2.COLOR_BGR2GRAY)
_, th = cv2.threshold(g, 220, 255, cv2.THRESH_BINARY)


con, h = cv2.findContours(th, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)

area = []

for c in con:
    m = cv2.moments(c)
    if m["m00"] != 0:  # Only proceed if the area is non-zero
        x = int(m["m10"] / m["m00"])
        y = int(m["m01"] / m["m00"])
        cv2.circle(i, (x, y), 2, (255, 0, 0), 1)  # Draw centroid

    cv2.drawContours(i, [c], -1, (255, 0, 0), 1)  # Draw individual contour
    a = cv2.contourArea(c)
    area.append(a)


    
m=.1*cv2.arcLength(c,True)
n=cv2.approxPolyDP(c,m,True)
h=cv2.convexHull(n)
x,y,w,h=cv2.boundingRect(h)
cv2.rectangle(i1,(x,y),(x+w,y+h),(0,255,0),3)

print("Contour Areas:", area)

cv2.imshow("Contours and Centroids", i)
cv2.imshow("Threshold", th)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
