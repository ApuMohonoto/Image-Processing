```python
#Ostu ThreShold ==Gives the threshold value between two peak point of the Histrogram
import cv2
import matplotlib.pyplot as plt
i=cv2.imread(r"R:\Ruponty\IMG_20250328_194659.jpg")
i1=cv2.resize(i,(500,500))
gray=cv2.cvtColor(i1,cv2.COLOR_BGR2GRAY)
hist=cv2.calcHist([gray],[0],None,[255],[0,255])
plt.plot(hist)
plt.show()
_,th=cv2.threshold(gray,100,255,cv2.THRESH_BINARY+cv2.THRESH_OTSU)
cv2.imshow("apu",gray)
cv2.imshow("apu",th)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
